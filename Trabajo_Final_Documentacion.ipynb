###INTRODUCCIÓN

Quoridor es un juego de estrategia abstracta de 2 a 4 jugadores diseñado por Mirko Marchesi y publicado por Gigamic Games. Quoridor recibió el premio Mensa Mind Game en 1997 y el juego del año en Estados Unidos, Francia, Canadá y Bélgica.

Quoridor también se basa en el juego anterior Blockade de Mirko Marchesi que se publicó en 1975. Marchesi también creó otra versión de este juego llamado Pinko Pallino que fue publicado en 1995 por Epta. Pinko Pallino era solo para 2 jugadores y se jugaba en un tablero de juego de 11 × 11 con un total de 42 paredes y reglas ligeramente diferentes.


##Reglas del juego

Quoridor se juega en un tablero de 81 casillas cuadradas (9x9). Cada jugador está representado por un peón que comienza en el espacio central de un borde del tablero (en un juego de dos jugadores, los peones comienzan uno frente al otro). El objetivo es ser el primer jugador en mover su peón a cualquier espacio en el lado opuesto del tablero desde el que comienza.

La característica distintiva de Quoridor son sus veinte paredes. Las paredes son piezas planas de dos espacios que se pueden colocar en la ranura que discurre entre los espacios. las paredes bloquean el camino de todos los peones, que deben rodearlos. Las paredes se dividen por igual entre los jugadores al comienzo del juego y, una vez colocadas, no se pueden mover ni quitar. En un turno, un jugador puede mover su peón o, si es posible, colocar una pared.

Los peones se pueden mover en cualquier espacio en ángulo recto (pero no en diagonal). Si está adyacente a otro peón, este puede saltar sobre ese peón. Si este es adyacente a una pared, el jugador puede moverse a cualquier espacio que esté inmediatamente adyacente (izquierda o derecha). No se pueden saltar varios peones. Las paredes no se pueden saltar.

Las paredes se pueden colocar directamente entre dos espacios, en cualquier ranura que no esté ocupada por una pared. Sin embargo, no se puede colocar una pared que corte el único camino restante de cualquier peón hacia el lado del tablero al que debe llegar.

###Definición de un algoritmo

•	Un algoritmo implica la descripción precisa de los pasos a seguir para alcanzar la solución de un problema dado. 

•	Por pasos se entiende el conjunto de acciones u operaciones que se efectúan sobre ciertos objetos.


         Algoritmo: Entrada              Proceso                       Salida

                                                    
###Características de un algoritmo

•	Un algoritmo debe poseer las siguientes características:

Precisión: Un algoritmo debe expresarse sin ambigüedad.

Determinismo: Todo algoritmo debe responder del modo antes las mismas condiciones.

Finito: La descripción de un algoritmo debe ser finita.



###Complejidad Algorítmica

•	La complejidad algorítmica representa la cantidad de recursos (temporales) que necesita un algoritmo para resolver un problema y por lo tanto permite determinar la eficiencia de dicho algoritmo.

•	Los criterios que se van a emplear para evaluar la complejidad algorítmica no proporcionan medidas absolutas sino medidas relativas al tamaño del problema.

###ESTADO DEL ARTE

Existen numerosos problemas que se pueden formular en términos de grafos. Esto nos brinda una percepción de lo que son las planificaciones de las tareas para encontrar las rutas de menor longitud para poder llegar a un objetivo en un tiempo específico. En este caso, implementaremos el juego de Quoridor. Por lo tanto, analizaremos tres algoritmos con la finalidad de encontrar cual es el más eficiente para la construcción del proyecto. Para esto tenemos como candidatos a los siguientes:


1.	Dijkstra
2.	Floyd-Warshall

###Dijkstra

El algoritmo de Dijkstra es aquél que encuentra el camino más corto desde un vértice al resto de vértices en un grafo, es un algoritmo ávido (voraz) y que sus aristas deben tener valores no negativos.

###Algoritmo de Floyd-Warshall

Es muy parecido al Algoritmo de Warshall, pero en el Algoritmo de Floyd las aristas tienen peso, además de encontrar el camino más corto Floyd lo representa de una manera más elegante y directa que el Algoritmo de Dijkstra. En este algoritmo se generan matrices consecutivas a partir de la primera, en cada paso se incorpora un nuevo vértice y se observa si con ese vértice se puede mejorar los caminos para ser más cortos.

###Conclusiones

1. El algoritmo de Floyd-Warshall es muchísimo más eficiente si se trata de encontrar el camino entre todas las parejas de nodos.


2. Por el contrario, el algoritmo de Dijkstra es mucho más eficiente si se encuentra el camino partiendo de un nodo hacia cualquier nodo del grafo.
